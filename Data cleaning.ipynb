```python
bar_heights = top5_rating['Rating'].tolist()

plt.bar(top5_rating['MovieName'], bar_heights)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
filtered_df = data[data['Year'] == 2019]
```

```python
plt.scatter(data.loc[data['Year'] == 2019, 'Genre'], filtered_df['Rating'])
plt.title('YEAR VS RATING')
```

```python
sns.lineplot(x=top5_hero_rating['Rating'],y=filtered_df['MovieName'])
plt.title('TOP 5 MOVIENAMES BASED ON HERORATING')
plt.xticks(rotation=0, fontsize=10)
plt.tight_layout()
plt.show()
```
